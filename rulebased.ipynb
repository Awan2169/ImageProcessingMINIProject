```python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def compute_thresholds(df):
    return df.select_dtypes(include=['number']).mean()

def rule_based_classifier(row, th):
    score = 0

    if row['laplacian_variance'] < th['laplacian_variance']:
        score += 2
    if row['edge_strength'] < th['edge_strength']:
        score += 2
    if row['fft_energy'] < th['fft_energy']:
        score += 1
    if row['noise_std'] < th['noise_std']:
        score += 1
    if row['entropy'] > th['entropy']:
        score += 1
    if row['contrast'] > th['contrast']:
        score += 1
    if row['homogeneity'] > th['homogeneity']:
        score += 1

    return 1 if score >5 else 0

def evaluate_rule_model(df):
    y_true = df['label'].map({'Real': 0, 'AI': 1})

    th = compute_thresholds(df)
    y_pred = df.apply(lambda row: rule_based_classifier(row, th), axis=1)

    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred)
    rec = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)

    print(f"Accuracy : {acc:.3f}")
    print(f"Precision: {prec:.3f}")
    print(f"Recall   : {rec:.3f}")
    print(f"F1 Score : {f1:.3f}")

df = pd.read_csv("Image_features_Dataset.csv")
evaluate_rule_model(df)
```
